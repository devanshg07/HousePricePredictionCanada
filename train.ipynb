```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib 
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python
df = pd.read_csv('housePrices.csv', encoding='latin1')
df.head()
```

```python
df = df.drop(['Address', 'Province', 'Population', 'Latitude', 'Longitude', 'Median_Family_Income'], axis=1)
df.head()
```

```python
df = pd.get_dummies(df, columns=['City'], dtype=int)
df.head()
```

```python
df.shape
```

```python
x = df.drop(['Price'],axis='columns')
x.head()
```

```python
x.shape
```

```python
y = df['Price']
y.head()
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(x_train,y_train)
lr_clf.score(x_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), x, y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X, y):
    algos = {
        'linear_regression': {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1, 2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['squared_error', 'friedman_mse'],
                'splitter': ['best', 'random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X, y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])

find_best_model_using_gridsearchcv(x, y)
```

```python
def predict_price(city, beds, baths):
    # Build the input vector with zeros
    x = np.zeros(len(x_train.columns))
    # Set the numeric features
    x[x_train.columns.get_loc('Number_Beds')] = beds
    x[x_train.columns.get_loc('Number_Baths')] = baths
    # Set the correct city column to 1 if it exists
    city_col = f'City_{city}'
    if city_col in x_train.columns:
        x[x_train.columns.get_loc(city_col)] = 1
    else:
        print(f"Warning: City '{city}' not found in training data. Prediction may be unreliable.")
    # Predict and return the price
    return lr_clf.predict([x])[0]
```

```python
predict_price('Halifax', 3, 2)
```
